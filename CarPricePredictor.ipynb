```python
X=df.drop(['Price','Engine volume','Cylinders','Doors','Wheel','Gear box type', 'Leather interior','ID','Levy','Color', 'Fuel type', 'Fuel type', 'Drive wheels', 'Airbags' ], axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['Model', 'Manufacturer','Category']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['Model', 'Manufacturer','Category']), remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
lr = LinearRegression()
```

```python
pipe=make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test, y_pred)
```

```python
score=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2, random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans, lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    score.append(r2_score(y_test,y_pred))
```

```python
np.argmax(score)
```
